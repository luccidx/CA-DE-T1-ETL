## Load Data 

```python
import pandas as pd

data = pd.read_csv(path + "/country_data.csv")
```

## Transform Data

```python
print(data.columns)
```

```python
data.name
```

```python
# Select columns of high importance
important_columns = [
    'name','capital','surface_area','gdp','gdp_per_capita','population','imports','exports','gdp_growth','life_expectancy_male','life_expectancy_female'
]

# Extract the important columns
important_data = data[important_columns]

print("Extracted important columns:")
important_data
```

```python
len(important_data)
```

```python
#Remove Duplicate records
important_data_unique = important_data.drop_duplicates()
len(important_data_unique)
```

```python
print("Total Number of Duplicate Columns Removed:", len(important_data) - len(important_data_unique))
```

```python
# Calculate the average life expectancy for each country
important_data_unique['average_life_expectancy'] = (important_data_unique['life_expectancy_male'] + important_data_unique['life_expectancy_female']) / 2

# Merge the average life expectancy back into the original DataFrame
important_data_newUnique = important_data_unique.copy()

# Display the first few rows of the updated DataFrame
print("Data with average life expectancy column:")
important_data_newUnique.head()
```

```python
important_data_newUnique.isnull().sum()
```

```python
# Remove rows with any missing values
cleaned_data = important_data_newUnique.dropna()
cleaned_data
```

```python
# Filter rows where gdp_per_capita is greater than 10,000
filtered_data = cleaned_data[cleaned_data['gdp_per_capita'] > 10000]

# Display the first few rows of the transformed dataset
print("Transformed dataset:")
filtered_data
```

```python
filtered_data.columns
```

```python
len(filtered_data)
```

## Load transformed data into a new CSV file

```python
# Define the path for the new CSV file
new_path = "./filtered_country_data.csv"
df = pd.DataFrame(filtered_data)
df.to_csv(new_path)
```

## Load the data to SQLite Database

```python
import sqlite3

conn = sqlite3.connect('country_gdp_data.db')
```

```python
create_table = "CREATE TABLE IF NOT EXISTS \
                country_gdp_data_table \
                (name TEXT, \
                capital TEXT, \
                surface_area REAL, \
                gdp REAL, \
                gdp_per_capita REAL,\
                 population REAL, \
                 imports REAL, \
                 exports REAL, \
                 gdp_growth REAL, \
                 life_expectancy_male REAL,\
                  life_expectancy_female REAL, \
                  average_life_expectancy REAL)"

cursor = conn.cursor()
cursor.execute(create_table)
```

```python
# Load the transformed data into the SQLite database
for row in df.itertuples(index=False):
    insert_data = """INSERT INTO country_gdp_data_table 
                     (name, capital, surface_area, gdp, gdp_per_capita, population, imports, exports, gdp_growth, life_expectancy_male, life_expectancy_female, average_life_expectancy) 
                     VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)"""
    cursor.execute(insert_data, (row.name, row.capital, row.surface_area, row.gdp, row.gdp_per_capita, row.population, row.imports, row.exports, row.gdp_growth, row.life_expectancy_male, row.life_expectancy_female, row.average_life_expectancy))
    
# Commit the transaction
conn.commit()
```

```python
check_table_data = "SELECT * FROM country_gdp_data_table"
cursor.execute(check_table_data)

# Fetch all results from the executed query
rows = cursor.fetchall()

# Print the results
for row in rows:
    print(row)
    
# Close the connection
conn.close()
```
